```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("predictions.csv")
num_classes = df.Labels.nunique()
confusion = np.zeros((num_classes, num_classes))
for _, r in df.iterrows():
    row = r.Labels
    col = r.Predictions
    confusion[row, col] += 1
print(confusion)
plt.imshow(confusion)
plt.colorbar(orientation='vertical')
plt.xlabel("Predicted classes")
plt.ylabel("True classes")
plt.title("Class Confusion Matrix")

print("Predicted Class Counts")
predicted = np.sum(confusion, axis=0)
print(predicted)

total = sum(predicted)

print("True Class Counts")
true = np.sum(confusion, axis=1)
print(true)


correct = 0
in1 = 0
in2 = 0
for i in range(num_classes):
    correct += confusion[i, i]
    s1 = max(0, i - 1)
    e1 = min(num_classes, i + 2)
    s2 = max(0, i - 2)
    e2 = min(num_classes, i + 3)
    in1 += np.sum(confusion[i, s1:e1])
    in2 += np.sum(confusion[i, s2:e2])
    
print("Percent Correct")
print(str((correct / total * 100)) + "%")

print("Percent Within 1 Classes")
print(str((in1 / total * 100)) + "%")

print("Percent Within 2 Classes")
print(str((in2 / total * 100)) + "%")
```
